```python
import cv2
import matplotlib.pyplot as plt
```

```python
cap = cv2.VideoCapture("/config/workspace/artifacts/tracker/output.mp4")
```

```python
for count in range(int(cap.get(cv2.CAP_PROP_FRAME_COUNT))):
    if count % 50 == 0:
        suc,frame = cap.read()
        if suc:
            plt.imshow(cv2.cvtColor(frame,cv2.COLOR_BGR2RGB))
            plt.show()
```
